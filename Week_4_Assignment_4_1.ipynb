```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('Titanic.csv')
```

1. Visualize the survival rate of passengers based on their ticket class (1st, 2nd, or 3rd class).
Were the passengers in higher classes more likely to survive?

```python
sns.barplot(x='class', y='survived', data=df)
plt.title('Survival rate based on ticket class') #adding a title for the chart
plt.xlabel('Ticket class') #labeling the x
plt.ylabel('Survival rate') #labeling the y
plt.xticks ([0,1,2], ['1st Class', '2nd Class', '3rd Class']) #making class numbers more clear
plt.ylim(0,1)
plt.show()
```

Explanation here:

2. Create a histogram to visualize the distribution of passenger ages. What is the age distribution like on the Titanic?

```python
sns.histplot(data=df, x='age', bins=30, kde=True) #making a histogram for the age column
plt.title('Age difference of Titanic Passengers') #adding a title
plt.xlabel ('Age') #labeling X
plt.ylabel ('Number of passengers') #labeling Y
plt.show()
```

Explanation here:

3. Generate a bar chart to compare the survival rates of male and female passengers. Did gender play a role in survival?

```python
surival_rate = df.groupby('sex') ['survived'].mean().reset_index() #finding the survival rate by sex
sns.barplot(x='sex', y='survived', data=surival_rate) #making the barplot with x axis as sex and y as survived
plt.title('Survival rate based on sex') #making the title for the plot
plt.xlabel('Sex') #labeling the X
plt.ylabel('Survaving rate') #labeling the Y
plt.ylim (0,1) #making the Y axis limit between 0 and 1, so it simulates 0 to 100 percent
plt.show()


```

Explanation here:

4. Visualize the distribution of ticket fares for each passenger class using box plots. How did ticket fares vary across different classes?

```python
sns.boxplot(x='class', y='fare', data=df) #making the box plot with X as the class and Y as the fare
plt.title('Ticket fare distribution based on class') #making the title of the boxplot
plt.xlabel('Passenger class') #labeling the X
plt.ylabel('Fare') #labeling the Y
plt.show()



```

Explanation here:

5. Create a grouped bar chart to visualize the survival rates of passengers across different age groups and genders. Were certain age groups or genders more likely to survive?

```python
age_bins = [0 , 12 , 18 , 35 , 60 , 100]
age_labels = ['Child (0-12)' , 'Teen (12-18)' , 'Young Adult (18-35)' , 'Adult (35-60)' , 'Senior (60+)'] #Defining the age groups
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
survival_ratess = df.groupby(['age_group', 'sex'])['survived'].mean().reset_index() #getting the survival rates based on group and gender
plt.figure(figsize = (10,6))
sns.barplot(x='age_group' , y='survived' , hue= 'sex', data=survival_ratess) #making the grouped bar chart
plt.title('Survival rate based on Age and Gender') #making the title for the bar chart
plt.xlabel('Age group') #Labeling the X axis
plt.ylabel('Survival rate') #Labeling the Y axis
plt.ylim(0,1) #making the y axis between 0 and 1 so its basically 0-100%
plt.show()



```
